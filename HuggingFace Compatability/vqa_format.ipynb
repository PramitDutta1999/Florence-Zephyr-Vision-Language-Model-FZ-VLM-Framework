```python
import pandas as pd
import os

# Load
df = pd.read_csv(" # train-png-slice-information")  # change if needed
```

```python
rows = []

QUESTION_TEMPLATES = {
    "location": "What is the location of the nodule?",
    "diameter": "What is the diameter of the nodule?",
    "attenuation": "What is the attenuation of the nodule?",
    "margin": "What is the margin of the nodule?"
}

for _, row in df.iterrows():
    pid = row.get("pid")
    image_path = row.get("png_path")

    # Skip invalid images
    if pd.isna(image_path) or not isinstance(image_path, str):
        continue

    # Ground truth values
    loc = row.get("sct_epi_loc")
    dia = row.get("sct_long_dia")
    att = row.get("sct_pre_att")
    mar = row.get("sct_margins")

    # Skip incomplete annotations
    if any(pd.isna(x) for x in [loc, dia, att, mar]):
        continue

    # One image → four QA rows
    rows.extend([
        {
            "pid": pid,
            "image": image_path,
            "question": QUESTION_TEMPLATES["location"],
            "answer": str(loc)
        },
        {
            "pid": pid,
            "image": image_path,
            "question": QUESTION_TEMPLATES["diameter"],
            "answer": str(dia)
        },
        {
            "pid": pid,
            "image": image_path,
            "question": QUESTION_TEMPLATES["attenuation"],
            "answer": str(att)
        },
        {
            "pid": pid,
            "image": image_path,
            "question": QUESTION_TEMPLATES["margin"],
            "answer": str(mar)
        }
    ])

# Final dataset
vqa_df = pd.DataFrame(rows)

# Save
vqa_df.to_csv("vqa_prompt_train.csv", index=False)
```
