# Combining Different Type of Work

```python
import json
import random
from pathlib import Path

random.seed(42)

longitudinal_file = "longitudinal_llm_zero_shot.jsonl"
followup_file = "followup_llm_zero_shot.jsonl"
nodule_file = "description_llm_zero_shot.jsonl"
output_file = "llm_input.jsonl"

# ---------- read jsonl ----------
def read_jsonl(path):
    data = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                data.append(json.loads(line))
    return data

longitudinal_data = read_jsonl(longitudinal_file)
followup_data = read_jsonl(followup_file)
nodule_data = read_jsonl(nodule_file)

print("Original longitudinal entries:", len(longitudinal_data))
print("Original followup entries:", len(followup_data))
print("Original nodule entries:", len(nodule_data))

# ---------- combine ----------
combined = longitudinal_data + followup_data + nodule_data

# ---------- randomize ----------
random.shuffle(combined)

print("Total combined entries:", len(combined))

# ---------- save ----------
with open(output_file, "w", encoding="utf-8") as f:
    for row in combined:
        f.write(json.dumps(row, ensure_ascii=False) + "\n")

print(f"Saved combined randomized file to {output_file}")
```
